```python
import numpy as np
import pandas as pd
```

```python
class LogisticRegression:
    def __init__(self,lr=0.01,iterations=1000):
        self.iteration=iterations
        self.lr=lr
        self.w=None
        self.b=None
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
        
    def fit(self,x,y):
        m,n=x.shape
        self.w=np.zeros(n)
        self.b=0
        for _ in range(self.iteration):
            linear_model=self.b+np.dot(x,self.w)
            y_pred=self.sigmoid(linear_model)

            # compute gradient
            dw=(1/m)*np.dot(x.T,(y_pred-y))
            db=(1/m)*np.sum(y_pred-y)

            self.w-=self.lr*dw
            self.b-=self.lr*db

    def predict_prob(self,x):
        linear_model=self.b+np.dot(x,self.w)
        return self.sigmoid(linear_model);
        
    def predict(self,x):
        y_predicted=self.predict_prob(x)
        return [1 if i>0.5 else 0 for i in y_predicted]
```

```python
# Sample dataset
X = np.array([[2,3], [1,1], [4,5], [6,7]])
y = np.array([0, 0, 1, 1])

# Create model
model = LogisticRegression(lr=0.1, iterations=1000)

# Train
model.fit(X, y)

# Predict
predictions = model.predict(X)

print("Predictions:", predictions)
```
