```python
class LinearRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.lr = learning_rate
        self.it = iterations
        self.w = None
        self.b = None
    def fit(self, X, y):
        m, n = len(X), len(X[0])   
        
        self.w = [0.0] * n   
        self.b = 0.0        

        for _ in range(self.it):
            dw = [0.0] * n
            db = 0.0
            for i in range(m):
                y_pred = self.b
                for j in range(n):
                    y_pred += self.w[j] * X[i][j]

                error = y_pred - y[i]
                for j in range(n):
                    dw[j] += error * X[i][j]

                db += error
            for j in range(n):
                self.w[j] -= self.lr * (2/m) * dw[j]

            self.b -= self.lr * (2/m) * db
    def predict(self, X):
        result = []
        for i in range(len(X)):
            y_pred = self.b
            for j in range(len(self.w)):
                y_pred += self.w[j] * X[i][j]
            result.append(y_pred)
        return result
```
